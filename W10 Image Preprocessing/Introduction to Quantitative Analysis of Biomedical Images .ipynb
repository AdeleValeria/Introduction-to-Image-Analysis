## Using Scikit to Read and Show Images

```python
# download images people.math.sc.edu/Burkardt/data/tif/tif.html
img = imread("at3_1m4_01.tif")
imshow(img)
```

```python
# crop image: take first 200 rows and 200 columns
img_small = img[0:200, 0:200]
imshow(img_small)
print(img_small.shape)
```

## Colourmapping in Matplotlib

```python
# plt.subplots allows plotting > 1 image side by side
fix, ax = plt.subplots(ncols = 3, nrows = 1, figsize = (10, 5))
ax[0].imshow(img, cmap = "viridis")
ax[1].imshow(img, cmap = "terrain")
ax[2].imshow(img, cmap = "spring")
plt.show()
```

## Image Histogram

```python
cell = imread("cameraman.tif")
imshow(cell, cmap = "gray")
plt.hist(cell.ravel(), bins = range(255))
plt.show()
```
